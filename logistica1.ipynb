```python
#Instalar las librerias
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

#Cargar los datos
Data = pd.read_csv("C:/Users/dacev/OneDrive/Escritorio/archive/datos.csv", sep=",")

#Mostrar los datos
Data

# Se realiza la selección de los datos /
# heartRate ritmo cardiaco y currentSmoker si es fumador  0 No 1 Si
Data[['heartRate','currentSmoker']].head()

#Generar grafica de puntos
Data[['heartRate','currentSmoker']].plot.scatter(x='heartRate',y='currentSmoker')

# Vamos a pintar una función logistica sobre estos datos

# pruebas de parametro
w = 0.09
b = -3.6

# puntos de la recta
x = np.linspace(0,Data['heartRate'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))


# grafica de la recta
Data.plot.scatter(x='heartRate',y='currentSmoker')
plt.plot(x, y, color='black')
plt.ylim(0,Data['currentSmoker'].max()*1.1)
plt.scatter(x, y, color='#A9E2F3')
# plt.grid()
plt.xlabel('Indice de Masa Corporal')
plt.ylabel('Diabetes 1:Positivo 0:Negativo')
plt.show()
```
